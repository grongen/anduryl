```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'
```

```python
quantiles = np.array([0.05, 0.5, 0.95])
estimates = np.array([5, 10, 20])
lower = 0.5 - 0.1 * 49.5
upper = 50 + 0.1 * 49.5
```

```python
bounds = np.r_[lower, estimates, upper]
full_quants = np.r_[0.0, quantiles, 1.0]

p = np.diff(full_quants)
np.log(upper - lower) + np.sum(p * np.log(p / (bounds[1:] - bounds[:-1])))
np.sum(p * np.log(p / (bounds[1:] - bounds[:-1])))
```

```python
xgrid = np.linspace(lower, upper, 10001)
dx = xgrid[1] - xgrid[0]
yinterp = np.interp(xgrid, bounds, full_quants)
fx = np.gradient(yinterp)

q = fx * dx

p = np.gradient(np.interp(xgrid, [lower, upper], [0.0, 1.0])) * dx

plt.plot(xgrid, fx)


np.sum(p * np.log(p / q))
```
